```python
def insertion_sort(A):
    for i in range(len(A)):
        cur_value = A[i]
        j = i-1
        while j >= 0 and A[j] > cur_value:
            A[j+1] = A[j]
            j -= 1
            
        A[j+1] = cur_value
    return A
```

```python
#this code is available if you wish to build different types of stable sort algotihms. Each one has its pros and cons.
def stable_sort(A, sort_type = "insertionsort"):
    if sort_type == "insertionsort":
        return insertion_sort(A)
```

```python
def bucket_sort(A, num_buckets = 10): 
    buckets = [] 
    for i in range(num_buckets): 
        buckets.append([]) 
        
    # Put array elements in different buckets according to a sellected key
    for j in A: 
        j_key = int(list(j)[0]) 
        buckets[j_key].append(list(j.values())[0]) 
    print(buckets)
    # applies a stable sort algorithm in each bucket
    # insertion sort algorithm is easier to read, but there are more efficient ones for this case, such as counting sort.
    for i in range(num_buckets): 
        buckets[i] = stable_sort(buckets[i], sort_type = "insertion") 
    # concatenate the result 
    k = 0
    for i in range(num_buckets): 
        for j in range(len(buckets[i])): 
            A[k] = buckets[i][j] 
            k += 1
    return A 
```

```python
# Python program for implementation of Radix Sort 

# A function to do counting sort of arr[] according to 
# the digit represented by exp. 
def countingSort(arr, exp1): 
    n = len(arr) 

    # The output array elements that will have sorted arr 
    output = [0] * (n) 
    
    # initialize count array as 0 
    C = [0] * (10) 
    
    # Store count of occurrences in count[] 
    for i in range(0, n): 
        index = int((arr[i]/exp1) )
        C[ (index)%10 ] += 1

    # Change count[i] so that count[i] now contains actual 
    # position of this digit in output array 
    for i in range(1,10): 
        C[i] += C[i-1] 
    
    # Build the output array 
    i = n-1
    while i>=0: 
        index = int((arr[i]/exp1) )
        output[ C[ (index)%10 ] - 1] = arr[i] 
        C[ (index)%10 ] -= 1
        i -= 1
    print(C)
    #print(output)
    # Copying the output array to arr[], 
    # so that arr now contains sorted numbers 
    i = 0
    for i in range(0,len(arr)): 
        arr[i] = output[i] 
```
